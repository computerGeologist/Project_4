```python
import pandas as pd
import numpy as np
from sklearn.decomposition import TruncatedSVD
from sklearn.decomposition import NMF
from sklearn.metrics.pairwise import cosine_similarity
torah = pd.read_csv("Torah_Chapters.csv")[['0']]

from nltk.tokenize import word_tokenize
from nltk.util import ngrams
import nltk
nltk.download('punkt')
nltk.download('stopwords')
```

```python
torah_words = [sentence for sentence in torah['0']]
twograms =list(ngrams(word_tokenize(torah_words[0]), 2))
twograms_words = [ngram[0]+ ngram[1] for ngram in twograms]
```

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.decomposition import LatentDirichletAllocation

stopwords = nltk.corpus.stopwords.words('english')
stopwords.extend(['thy', 'thou', "thee", "god"])

tf_vectorizer = CountVectorizer(stop_words = stopwords, ngram_range=(1,2))
chapters_tf = tf_vectorizer.fit_transform(torah_words)
ch_tf= pd.DataFrame(chapters_tf.toarray(),columns=tf_vectorizer.get_feature_names())
```

```python
components = 7
lsa = TruncatedSVD(components)
doc_topic = lsa.fit_transform(chapters_tf)
lsa.explained_variance_ratio_
```

```python
topic_word = pd.DataFrame(lsa.components_.round(3),
             index = ["components_"+str(i+1) for i in range(components)],
             columns = tf_vectorizer.get_feature_names())
topic_word
```

```python
def display_topics(model, feature_names, no_top_words, topic_names=None):
    for ix, topic in enumerate(model.components_):
        if not topic_names or not topic_names[ix]:
            print("\nTopic ", ix)
        else:
            print("\nTopic: '",topic_names[ix],"'")
        print(", ".join([feature_names[i]
                        for i in topic.argsort()[:-no_top_words - 1:-1]]))
```

```python
display_topics(lsa, tf_vectorizer.get_feature_names(), 10)
```
